How are these categories defined?

* **Air Carrier:** The cause of the cancellation or delay was due to circumstances within the airline's control (e.g. maintenance or crew problems, aircraft cleaning, baggage loading, fueling, etc.).
* **Extreme Weather:** Significant meteorological conditions (actual or forecasted) that, in the judgment of the carrier, delays or prevents the operation of a flight such as tornado, blizzard or hurricane.
* **National Aviation System (NAS):** Delays and cancellations attributable to the national aviation system that refer to a broad set of conditions, such as non-extreme weather conditions, airport operations, heavy traffic volume, and air traffic control.
* **Late-arriving aircraft:** A previous flight with same aircraft arrived late, causing the present flight to depart late.
* **Security:** Delays or cancellations caused by evacuation of a terminal or concourse, re-boarding of aircraft because of security breach, inoperative screening equipment and/or long lines in excess of 29 minutes at screening areas.

source: https://www.bts.gov/topics/airlines-and-airports/understanding-reporting-causes-flight-delays-and-cancellations

I will plot one of these variables in order to have a better understanding:

```python
fig, (ax1, ax2) = plt.subplots(figsize = (16, 5), nrows = 1, ncols = 2)

fig.suptitle('ORIGINAL', fontsize=16)

plt.subplot(1, 2, 1)
plt.scatter(flights.DepDelay, flights.CarrierDelay)
plt.title('DepDelay vs. CarrierDelay - Original')

plt.subplot(1, 2, 2)
plt.hist(flights.CarrierDelay, bins = 1000)
plt.xlim(0, 30) # median is 2, 75% quartile is 21
plt.title('Histogram Zoom CarrierDelay - Original')

plt.show()
```

There are quite a lot of NaN's in the last five columns (around 35%) so dropping these rows is not an option. I think is better to fill them with the mean value rather than 0. Interpolation wolud be another way, let's try both:

*Replacing NaN's with the mean*

```python
flights_mean = flights.copy() # copy dataframe

for i in flights_mean.columns[flights_mean.isnull().any(axis = 'index')]:    
    flights_mean[i].fillna(flights_mean[i].mean(), inplace = True)
```

```python
fig, (ax1, ax2) = plt.subplots(figsize = (16, 5), nrows = 1, ncols = 2)

plt.subplot(1, 2, 1)
plt.scatter(flights_mean.DepDelay, flights_mean.CarrierDelay)
plt.title('DepDelay vs. CarrierDelay')

plt.subplot(1, 2, 2)
plt.hist(flights_mean.CarrierDelay, bins = 1000)
plt.xlim(0, 30) # median is 2, 75% quartile is 21
plt.title('Histogram Zoom - Carrier Delay')

fig.suptitle('NAN-MEAN', fontsize=16)

plt.show()
```

```python
flights_mean.isnull().sum().sum()
```

As expected, because the number of NaN's we have, there is a substantial difference in the distribution of the variable. We have a bimodal distribution.

*Replacing NaN's by interpolation*

```python
flights_interpol = flights.copy() # copy dataframe

for i in flights_interpol.columns[flights_interpol.isnull().any(axis = 'index')]:
    flights_interpol[i].interpolate(method='linear', axis = 'index', inplace = True)
```

```python
fig, (ax1, ax2) = plt.subplots(figsize = (16, 5), nrows = 1, ncols = 2)

plt.subplot(1, 2, 1)
plt.scatter(flights_interpol.DepDelay, flights_interpol.CarrierDelay)
plt.title('DepDelay vs. CarrierDelay')

plt.subplot(1, 2, 2)
plt.hist(flights_interpol.CarrierDelay, bins = 1000)
plt.xlim(0, 30) # median is 2, 75% quartile is 21
plt.title('Histogram Zoom')

fig.suptitle('NAN - INTERPOLATE', fontsize=16)

plt.show()
```

### **Variables Type Classification**

Let's continue by classifying the variables type correctly.

```python
flights['Cancelled'] = flights['Cancelled'].replace([1,0],["Cancelled","Not Cancelled"])
flights['Diverted'] = flights['Diverted'].replace([1,0],["Diverted","Not Diverted"])
```

We can join first 3 columns into datetime:

```python
date = flights.rename(columns = {'DayofMonth':'Day'}) 
date = date[['Year','Month','Day']] 
flights.insert(0, 'Date', pd.to_datetime(date))
flights = flights.drop(['Year','Month','DayofMonth'], axis=1)
flights['DayOfWeek']= flights['DayOfWeek'].replace({1: "Monday", 2: "Tuesday", 3: "Wednesday", 
                                                    4: "Thursday", 5: "Friday", 6: 'Saturday', 7:'Sunday'})
```

For categorical variables:

```python
cat_var = ['DayOfWeek', 'UniqueCarrier', 'FlightNum', 'TailNum', 'Origin', 'Dest', 'CancellationCode', 'Cancelled', 'Diverted']
cancel_code = {"A" : "Carrier", "B" : "Weather", "C" : "NAS", "D" : "Security", "N" : "None" }

flights['CancellationCode'] = flights['CancellationCode'].replace(cancel_code)
flights[cat_var] = flights[cat_var].astype('category')
```

For numerical variables:

```python
num_cols = flights._get_numeric_data().columns

flights[num_cols] = flights[num_cols].astype('int64')
```

### **Standarize numerical data**

Previously we definden **num_cols** with a list of columns with numerical data.

Also we saw that there are quite a lot of outliers so it is a good idea to use RobustScaler from SKLearn library to scale data taking into consideration these outliers.

```python
from sklearn.preprocessing import RobustScaler
```

```python
flights_st = pd.DataFrame(RobustScaler(quantile_range = (25.0, 75.0)).fit_transform(flights[num_cols]))
flights_st.columns = flights[num_cols].columns
```

```python
flights_st
```

```python
flights_st.describe().transpose()
```

### **Principal Component Analysis with standarized data**

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(.95) #95% of the variance is retained
x = pca.fit_transform(flights_st)
flights_PCAst = pd.DataFrame(data = x)
flights_PCAst
```

How much variance explains each PCA column for train dataset:

```python
pca_explst = pca.explained_variance_ratio_.round(3) # how much variance exlains each PCA column
pca_explst
```

We went from having 19 columns to only 2 with Principal component analysis and PCA1 explains 93,6% of the variance, not bad.

```python
pca_biplot = pd.concat([flights_PCAst, flights.DayOfWeek], axis = 'columns')
pca_biplot = pca_biplot.sample(100000)
pca_biplot
```

```python
fig, ax = plt.subplots(figsize = (16, 5))
sns.scatterplot(x = pca_biplot[0], y = pca_biplot[1], hue = pca_biplot['DayOfWeek'])
plt.title('PCA Biplot')
plt.show()
```

We have variables with very strong linear correlation . If we take a closer look at them, for example AirTime and Distance, it's quite obvious that these to variables are very much correlated.
